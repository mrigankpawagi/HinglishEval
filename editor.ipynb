````python
import os
import re

# create a new folder named doc2
# os.mkdir('doc2')

# set the folder path
folder_path = 'doctests'

# loop through the files in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.md'):
        # set the file paths
        file_path = os.path.join(folder_path, file_name)
        new_file_path = os.path.join('doc2', file_name)
        with open(file_path, 'r') as f:
            lines = f.readlines()
            print(lines)
        with open(new_file_path, 'w') as f:
            i = 0
            while i < len(lines):
                if lines[i].lstrip().startswith('assert'):
                    match = re.match(r'assert\s+(.*)\s+==\s+(.*)', lines[i].lstrip())
                    if match:
                        f.write(f'>>> {match.group(1)}\n{match.group(2)}\n')
                        i += 1
                        while i < len(lines) and not lines[i].lstrip().startswith('assert'):
                            f.write(lines[i])
                            i += 1
                    
                    elif not lines[i].lstrip().startswith('assert'):
                        f.write(lines[i])
                        print("else1")
                        print(lines[i])
                        i += 1
                        continue

                    # i += 1
                else:
                    f.write(lines[i])
                    print("else2")
                    print(lines[i])
                    i += 1
                    continue
````
